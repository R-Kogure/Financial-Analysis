## 1. Variation of Sample

```python
import pandas as pd
import numpy as np
from scipy.stats import norm
%matplotlib inline
```

```python
a_sample = pd.DataFrame(np.random.normal(10,5,size= 30)) #Î¼ = 10, SD = 5
print('sample mean is ', a_sample[0].mean())
print('sample SD is ', a_sample[0].std(ddof=1))
```

## 2. Empirical Distribution of mean

```python
mean_list = []
trial = 10000

for _ in range(trial):
    a_sample = pd.DataFrame(np.random.normal(10, 5, size = 30))
    mean_list.append(a_sample[0].mean())
```

```python
# plot
collection = pd.DataFrame()
collection['MeanList'] = mean_list

collection['MeanList'].hist(bins = 100,normed = 1, figsize = (16,9)) #normed = 1 : æ­£è¦å
```

## 3. Sampling from arbritary distribution
What is Central Limit Theorem ?


```python
sample_size = 100
trial = 10000
SampleMeanList = []
NotNormal = pd.DataFrame([0, 0, 1, 1, 1])
```

```python
for _ in range(trial):
    tmp = NotNormal[0].sample(sample_size, replace = True)
    SampleMeanList.append(tmp.mean())

a_col = pd.DataFrame()
a_col['MeanList'] = SampleMeanList
a_col['MeanList'].hist(bins = 100,normed = 1, figsize = (16,9)) #normed = 1 : æ­£è¦å
```

## 4. Conclusion
Even if the population is not normal distribution, the sample size is larger enough, the distribution of sample mean is approximately normal
